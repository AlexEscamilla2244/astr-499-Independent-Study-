```python
import yt
import astropy.units as u
%matplotlib inline 
import matplotlib.pyplot as plt
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
import h5py
from glob import glob
from yt.fields.api import ValidateParameter
```

```python
unit_base = {
    "length": (1.0, "pc"),
    "velocity": (100, "cm/s"),
    "mass": (1.0, "Msun"),
}
data = yt.load('../data/snapshot_200.hdf5', unit_base=unit_base)
```

```python
def RadVelocity(field,data):
    ad = data.all_data()
    if data == ('bulk_velocity'):
        bv = data.get_field_paramater('bulk_velocity').in_units('cm/s')
    else:
        bv = data.array(np.zeros(3), 'cm/s')
    xv = data['velocity_x'] - bv[0]
    yv = data['velocity_y'] - bv[1]
    zv = data['velocity_z'] - bv[2]
    center = data.get_field_paramter('center')
    x_hat = data['x'] - center[0]
    y_hat = data['y'] - center[1]
    z_hat = data['z'] - center[2]
    r = np.sqrt(x_hat**2+y_hat**2+z_hat**2)
    x_hat /= r
    y_hat /= r
    z_hat /= r
    return  xv*x_hat + yv*y_hat + zv*z_hat
yt.add_field('radial_velocity', function= RadVelocity, sampling_type='particle', units='km/s',
              force_override=True)

# radial momentum multiply rad velocity with gas density
```
